### Starter code for HW 2

This notebook goes over a simple way to approximate the solution to a PDE, or ODE, using neural networks.  It treats the discretized PDE as a high dimensional ODE and learns a timestepper, abliet not one that generalizes to other initial conditions.  To run, you'll need to install tensorflow.

https://www.tensorflow.org/install/

```python
%pylab inline

import numpy as np
import tensorflow as tf

from scipy.integrate import odeint
from numpy.fft import fft, ifft, fftfreq
```

First get some data to try training on.  Running the next block of code with simulate Burgers' equation on a cylinder with a few different initial conditions and store the results.  We can then try to train a neural network to recreate these results starting from the initial condition alone and see how it performs on an unseen initial condition.

```python
def burgers_rhs(u, t, params):
    k = params
    deriv = -u*ifft(1j*k*fft(u)) + 0.1*ifft(-k**2*fft(u))
    return real(deriv)

# Set size of grid
n = 256
m = 257

# Set x,y,k
x = np.linspace(-8,8,n+1)[:-1];   dx = x[1]-x[0]
t = np.linspace(0,10,m);          dt = t[1]-t[0]
k = 2*np.pi*fftfreq(n, d = dx)

# A few different random initial conditions
num_sols = 11
U0 = [np.exp(-(x+1)**2) + np.sum([0.25**i*np.random.randn()*np.sin(i*np.pi*x/8+np.random.randn()) \
                                  for i in range(1,5)], axis = 0) for j in range(num_sols)]
    
# Solve on periodic domain
params = (k)
U = [odeint(burgers_rhs, U0[j], t, args=(params,)).T for j in range(num_sols)]

figure(figsize = (12,4))

subplot(1,2,1)
for u0 in U0: plot(u0)
title('Initial Conditions')

subplot(1,2,2); pcolor(U[0].T, cmap=cm.coolwarm); xlim([0,n]); ylim([0,m])
title('Example solution')
```

Now set up a simple neural network to try to act as a timestepper.  

1.  dense_layer takes as input a vector $x$, multiplies by a weight matrix $W$, and adds bias $b$.  If linear is set to True, it will return $Wx+b$.  Otherwise it clips the negative component of the resulting vector and returns $(Wx+b)_+$. 

2. simple_net applies dense layer to a vector several times, with the last application being linear.  To run this, we'll need to pass in a lists of TensorFlow variables for the weights and biases, to be defined later.

3. minibatch_feeder is a generator.  It is initialized by passing in a list of solutions to a PDE. The generator can returns pairs of matrices $u_0$ and $u_1$, each of size $n$ x minibatch size.  The relation between $u_0$ and $u_1$ is that each column of $u_0$, if propogated forward one timestep via the above code for Burgers, will give the corresponding column in $u_1$

```python
def dense_layer(x, W, b, linear = False):
    
    x = tf.matmul(W,x)
    x = tf.add(x,b)
               
    if linear: return x
    else: return tf.nn.relu(x)

def simple_net(x, weights, biases):
    
    layers = [x]
    
    for l in range(len(weights)-1):
        layers.append(dense_layer(layers[l], weights[l], biases[l]))

    out = dense_layer(layers[-1], weights[-1], biases[-1], linear = True)
    
    return out

def minibatch_feeder(U, minibatch_size, maxit = 1e8):
    
    num_sols = len(U)
    n,m = U[0].shape
    queue = list(np.random.permutation(num_sols*(m-1)))
    
    U0 = np.hstack([u[:,:-1] for u in U])
    U1 = np.hstack([u[:,1:] for u in U])
    
    for i in range(int(maxit)):
        
        # Make sure queue is sufficiently long to take away minibatch_size
        if len(queue) < minibatch_size:
            queue = queue + list(np.random.permutation(num_sols*(m-1)))
            
        # Take top entries from queue
        minibatch = queue[:minibatch_size]
        
        # Remove minibatch from queue
        queue = queue[minibatch_size:]
        
        # Get slices of u to pass back
        u0_minibatch = np.hstack([U0[:,j].reshape(n,1) for j in minibatch])
        u1_minibatch = np.hstack([U1[:,j].reshape(n,1) for j in minibatch])        
                
        yield (u0_minibatch, u1_minibatch)
```

```python
#Initialize a generator to get random batches of training data

feeder = minibatch_feeder(U, 5)
```

Try running this block of code several times.  You should see a random set of snapshots contained in the list of solutions U, followed by very slightly different snapshots (one step forward) and the difference betwen the two.

```python
u0,u1 = next(feeder)

print('u0 shape:', u0.shape) # 5 snapshots on u, each in R^n
print('u1 shape:', u1.shape) # same as u0 but 1 step forward in time

figure(figsize = (16,4))
subplot(1,3,1); plot(u0); title('u0'); xlim([0,n])
subplot(1,3,2); plot(u1); title('u1'); xlim([0,n])
subplot(1,3,3); plot(u1-u0); title('u1-u0'); xlim([0,n])
```

Running this cell erases any relations previously defined bewteen tensorflow variables and initializes a TF session.  Variables, placeholders (to be defined), and the relations between them need to be defined all within the same tensorflow session.

```python
tf.reset_default_graph()
sess = tf.InteractiveSession()
```

This sets up a set of weights and biases which we can pass as arguments into simple_net.  These are variables, so when we call on tensorflow to train the neural network given some dataset, their values will change.  Feel free to play with the first two lines here and everything will still work.

```python
n_hidden = 5         # number of hidden layers in the neural network
size_hidden = 512    # size of each hidden layer

###################################################################

layer_sizes = [n] + [size_hidden for _ in range(n_hidden)] + [n]
num_layers = len(layer_sizes)

weights = []
biases = []

for j in range(1,num_layers):
    
    weights.append(tf.get_variable("W"+str(j), [layer_sizes[j],layer_sizes[j-1]], \
                                   initializer = tf.contrib.layers.xavier_initializer(seed = 1)))
    
    biases.append(tf.get_variable("b"+str(j), [layer_sizes[j],1], initializer = tf.zeros_initializer()))
```

Now we define a placeholder u0 to hold values for the input of the neural net, and a few variables delta_u and u_1 which will be the change in u and prediction for u at the following timestep.  We also define a placeholder for the true value at the next timestep.  


For now, the actual values of all these variables are not defined.  Placeholders are used to define inputs to a model where we want to able to pass in whatever value we want.  Now that the weights and biases are defined, we can pass in a value of u0 and evaluate the resulting prediction.  During training, we will pass in values of u0 and u1_true, then look at the difference between u1_pred and u1_true to see how the model is doing.

```python
u0 = tf.placeholder(tf.float32, [n,None], name = "u0")
delta_u = simple_net(u0, weights, biases)
u1_pred = u0+delta_u

u1_true = tf.placeholder(tf.float32, [n,None], name = "u1_true")
```

Now we can set up a cost function, which will be minimized.  We want predictions to match the truth, but may also want some regularization.  The weights regularizer term is a natural result of putting a Gaussian distributed prior on the weights of the neural network, if we also assume the error u1_true-u1_pred is Gaussian.  This is a neural network equivalent of ridge regression.  The exact degree of regularization is specified by beta, which for now is a placeholder.

```python
beta = tf.placeholder(tf.float32, name = "beta")

# does the prediction match the truth?
model_fidelity = tf.nn.l2_loss(u1_true-u1_pred) 

# is our neural network overly complicated (overfitting)?
weights_regularizer = beta*tf.reduce_mean([tf.nn.l2_loss(W) for W in weights]) 

cost = model_fidelity + beta*tf.reduce_mean([tf.nn.l2_loss(W) for W in weights])
```

Specify an optimizer.  Adam is a standard method for neural networks based on stochastic gradient descent.  If you're interested, https://arxiv.org/abs/1412.6980

```python
# learning rate (i.e. step size) for Adam
lr = tf.placeholder(tf.float32, name = "lr")

optimizer = tf.train.AdamOptimizer(learning_rate=lr).minimize(cost)
```

Before doing anything with all the variables we've just defined we have to initialize them.  Try running the cell two below this prior to running the global variables initializer.  You should get an error.  Then initialize and try again.  It will show an initial condition for one of our datasets, and the untrained neural networks prediction for the next timestep.  The difference should at this point just look like noise.

```python
init = tf.global_variables_initializer()
sess.run(init)
```

```python
# this first line of code tells tensorflow to estimate u(x,t1) given that u(x,t0) is the IC for U[0]
# feed_dict tells tensorflow what values to assign the placeholders that are necesarry to find u1_pred

nn_prediction = u1_pred.eval(feed_dict = {u0 : U[0][:,0].reshape(256,1)})

plot(U[0][:,1].reshape(256,1), label = 'inital_state')
plot(nn_prediction, label = 'prediction')
xlim([0,n])
```

Try training on a single run.  Note that I'm initializing the minibatch_feeder with just one run of Burgers, so we're seeing if the neural network can learn to recreate that one solution.

1. num_epochs is the number of times Adam (like stochastic gradient descent) will pass over the entire dataset

2. minibatch_size if the number of time slices the algorithm will see at each step of the optimization.

```python
num_epochs = 300
minibatch_size = 32

####################################################################################

batches_per_epoch = int((m-1)/minibatch_size)
feeder = minibatch_feeder([U[0]], minibatch_size)
epoch_costs = np.zeros(num_epochs)

# Pass over the data num_epochs times
for epoch in range(num_epochs):

    for batch in range(batches_per_epoch):

        # Select a minibatch i.e. a subset of the data used to evaluate hpw our model is doing 
        # and approximate the derivative of the cost function
        (u0_minibatch, u1_minibatch) = next(feeder)

        # update the weights and biases using this minibatch
        # we also specify values for the step size and regularization parameter
        _ , minibatch_cost = sess.run([optimizer, cost], feed_dict={u0: u0_minibatch, 
                                                                    u1_true: u1_minibatch,
                                                                    lr: 0.01*0.99**epoch, 
                                                                    beta: 1e-3})
        
        # record the cost function evaluated on this minibatch
        epoch_costs[epoch] += minibatch_cost / batches_per_epoch

    # Print the cost every epoch
    if (epoch+1)%25 == 0 or epoch == 0: print ("Cost after epoch %i: %f" % (epoch+1, epoch_costs[epoch]))
```

```python
semilogy(epoch_costs)
xlabel('Epoch')
ylabel('Cost')
```

See how well it does on the training data.  Starting from the same IC, does the solution via NN look like the true solution?
